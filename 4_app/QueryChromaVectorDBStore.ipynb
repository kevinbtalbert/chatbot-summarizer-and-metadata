```python
## Initialize a connection to the running Chroma DB server
import chromadb
import os
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings.sentence_transformer import SentenceTransformerEmbeddings

## Use the following line to connect from within CML
chroma_client = chromadb.PersistentClient(path="/home/cdsw/chroma_db")

embedding_function = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")


#chroma_client = chromadb.HttpClient(host=HOST)
langchain_chroma = Chroma(
    client=chroma_client,
    collection_name=os.getenv("COLLECTION_NAME"),
    embedding_function=embedding_function,
)
```

```python
# query it
query = "What is a plane"
docs = langchain_chroma.similarity_search(query)

# print results
print(docs[0])
print("-------------")
print(docs[0].page_content)
```
